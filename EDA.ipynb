#   EDA - Exploratory Data Analysis for spotify dataset

#### Exploratory Data Analysis (EDA) means simply looking at your data for the first time to understand whatâ€™s inside, find mistakes, and see basic patterns.

### steps:
* Import Libraries
* Load data
* checking rows,columns,dtypes,shape,info,shape.
* identified duplicates ,missing values.
* Univariate analysis helps understand the distribution of each variable individually.
* Bivariate analysis helps us see how two variables are connected or affect each other. 
* A correlation matrix shows how strongly different numerical features are related to each other.
* A boxplot helps us spot unusual values and understand how the data is spread.

## IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### These are the main libraries used in EDA. pandas for data manipulation, numpy for numerical operations, matplotlib and seaborn for visualization

## Load Dataset

```python
df = pd.read_csv("spotify_analysis_dataset.csv")
```

## Rows

```python
df.head()
```

#### Displays the first 5 rows of the dataset. Helps us quickly understand the structure, column names, and sample values.

## Dataset Shape

```python
df.shape
```

#### Returns the number of rows and columns in the dataset. Helps us understand the dataset size and dimensionality.

## Column

```python
df.columns
```

#### Lists all the column names present in the dataset. Helps us verify feature names and check if all expected columns exist

## Data Types

```python
df.dtypes
```

#### Shows the data type (int, float, object, etc.) of each column.Helps identify numerical, categorical, and datetime variables

## Dataset Information

```python
df.info()
```

#### Provides a summary of the dataset, including Number of non-null values Data types Memory usage Helps detect missing data and confirm data types.

## Missing Values Count

```python
df.isnull().sum()
```

#### Counts the number of missing values in each column. Helps identify features needing cleaning or imputation

## Check Duplicate Rows

```python
df.duplicated().sum()
```

#### Counts how many duplicate rows exist in the dataset.

## Statistical Summary

```python
df.describe()
```

#### Generates summary statistics for numerical columns, including Mean,Median,Standard deviation ,Minimum & maximum

## Univariate analysis

```python
sns.histplot(df['popularity'], kde=True , color='pink', bins=30)
plt.show()
```

#### Shows the distribution of the popularity variable.Helps identify whether the values are Skewed,Uniform ,Normally distributed .The KDE curve smooths the distribution for better understanding

## Bivariate analysis

```python
sns.scatterplot(x='danceability', y='popularity', data=df, hue='energy', alpha=0.5, palette='viridis')
plt.show()
```

#### Visualizes the relationship between danceability and popularity.Helps identify whether higher danceability leads to higher popularity

## Correlation Matrix

```python
corr = df[['energy', 'tempo']].corr()
sns.heatmap(corr, annot=True, cmap="viridis")
plt.show()
```

#### Shows the correlation between energy and tempo.Helps identify whether increasing tempo affects the energy level of songs.

## Boxplot to Detect Outliers

```python
sns.boxplot(data=df.select_dtypes(include=np.number), palette="Set2")
plt.xticks(rotation=45)
plt.show()

```

### The boxplot shows that duration_ms has a very large range, while all other features have very small values and appear compressed.
